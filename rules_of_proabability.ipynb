```python
set1 = {'A', 'B','C','D','E'}
set2  = {'B','D','V','X','Y','Z'}
```

```python
union = set1.union(set2)
```

```python
total_guests = list(union)
```

```python
print("Total guests to be invited in party are :",len(total_guests))
```

```python
print("Guest List :", total_guests)
```

```python
union1 = set1.intersection(set2)
```

```python
total_guests1= list(union1)
```

```python
print("Total guests to be invited in party are :",len(total_guests1))
```

```python
print("Guest List :", total_guests1)
```
